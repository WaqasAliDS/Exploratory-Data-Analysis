```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Wine_Quality.csv')
df.head()
```

```python
columns = df.columns
columns
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
missing_rows = df[df.isnull().any(axis=1)]
missing_rows
```

```python
missing_values_column = df.columns[df.isnull().any()].tolist()
missing_values_column
```

```python
plt.figure(figsize=(12, 8))

for i, column in enumerate(missing_values_column, 1):
    plt.subplot(3, 3, i)
    df.boxplot(column=column)
    plt.title(f"Box Plot for {column}")

plt.tight_layout()

plt.show()
```

```python
column_means = df[missing_values_column].mean()
```

```python
column_means
```

```python
df[missing_values_column] = df[missing_values_column].fillna(column_means)
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
df["type"].unique()
```

```python
type_counts = df['type'].value_counts()
```

```python
colors = ['silver', 'red']

plt.bar(type_counts.index, type_counts.values, color=colors)

for i, count in enumerate(type_counts.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.xlabel('Type of Wine')
plt.ylabel('Quantity')
plt.title('Types of Wine')
plt.show()
```

```python
df_red = df[df['type'] == 'red']
df_red
```

```python
df_white = df[df['type'] == 'white']
df_white
```

## All Types of Wines

```python
columns_except_type = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol',
       'quality']
```

```python
def correlation(headers):
    correlation_matrix = df.corr()

    correlation_with_others = correlation_matrix[headers]

    print(f"Correlation of {headers} with each columns{correlation_with_others}")
    return
```

```python
correlation(columns_except_type)
```

```python
correlation_matrix = df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.scatter(df['free sulfur dioxide'], df['total sulfur dioxide'], marker='o', color='red')

plt.xlabel('free sulfur dioxide')
plt.ylabel('total sulfur dioxide')
plt.title('Scatter Plot of free sulfur dioxide vs total sulfur dioxide')
plt.show()
```

```python
plt.scatter(df['density'], df['residual sugar'], marker='o', color='orange')

plt.xlabel('Density')
plt.ylabel('Residual Sugar')
plt.title('Scatter Plot of Density vs Residual Sugar')
plt.show()
```

```python
plt.scatter(df['density'], df['alcohol'], marker='o', color='blue')

plt.xlabel('Density')
plt.ylabel('alcohol')
plt.title('Scatter Plot of Density vs alcohol')
plt.show()
```
