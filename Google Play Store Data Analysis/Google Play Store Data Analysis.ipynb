```python
mean_rating = round(df["Rating"].mean(),1)
mean_rating
```

```python
df["Rating"].fillna(mean_rating,inplace = True)
```

```python
missing_values = df[df.isnull().any(axis = 1)]
missing_values
```

```python
row_to_shift = df.iloc[10472, :]
row_to_shift
```

```python
## Shift the values one column to the right
row_to_shift.iloc[2:] = row_to_shift.iloc[1:-1].values
row_to_shift
```

```python
# Update the DataFrame with the shifted row
df.iloc[10472, :] = row_to_shift
```

```python
missing_values = df[df.isnull().any(axis = 1)]
missing_values
```

```python
df['Category'] = df['Category'].replace('1.9', 'LIFESTYLE')
df['Genres'] = df['Genres'].fillna('Lifestyle') 
```

```python
df['Type'].fillna('Free', inplace = True)
```

```python
df['Android Ver'] = df['Android Ver'].fillna(df['Current Ver'])
```

```python
missing_values = df[df.isnull().any(axis = 1)]
missing_values
```

```python
def format_android_version(android_ver):
    parts = android_ver.split()
    return parts[0]
```

```python
df['Android Ver (and up)'] = df['Android Ver'].apply(format_android_version)
df
```

```python
df['Current Ver'] = df['Current Ver'].fillna(df['Android Ver (and up)'])
```

```python
missing_values = df[df.isnull().any(axis = 1)]
missing_values
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
df['Reviews'] = df['Reviews'].astype(int)
```

```python
df.rename(columns={'Installs': 'Installs (+)'}, inplace=True)
df['Installs (+)'] = df['Installs (+)'].str.replace('+', '')
```

```python
df['Installs (+)'] = df['Installs (+)'].str.replace('+', '')
```

```python
df['Installs (+)'] = df['Installs (+)'].str.replace(',', '')
```

```python

df['Rating'] = df['Rating'].astype(float)
```

```python
df['Installs (+)'] = df['Installs (+)'].astype(int)
```

```python
df['Price'] = df['Price'].str.replace('$', '')
df['Price'] = df['Price'].astype(float)
```

```python
df.dtypes
```

```python
df['Size'].replace('Varies with device', np.nan, inplace = True )
```

```python
df.isnull().sum()
```

```python
df.rename(columns={'Size': 'Size (M)'}, inplace=True)
```

```python
def convert_to_mb(size):
    if pd.notnull(size):
        if 'k' in size:
            return float(size.replace('k', '')) / 1000
        elif 'M' in size:
            return float(size.replace('M', ''))
    return None

df['Size (M)'] = df['Size (M)'].apply(convert_to_mb)
```

```python
df.head()
```

```python
mean_size_by_category = df.groupby('Category')['Size (M)'].mean()
mean_size_by_category
```

```python
df['Size (M)'] = df.apply(lambda row: mean_size_by_category[row['Category']] if pd.isnull(row['Size (M)']) else row['Size (M)'], axis=1)
```

```python
sum_reviews_by_category = df.groupby('Category')['Reviews'].sum()
sum_reviews_by_category
```

```python
order = df.groupby('Category')['Reviews'].sum().sort_values().index
order
```

```python
plt.figure(figsize=(12, 6))
ax = sns.barplot(x='Category', y='Reviews', data=df, estimator=sum, ci=None, order=order)
plt.xticks(rotation=90)
plt.xlabel('Category')
plt.ylabel('Total Reviews (Log Scale)')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.title('Number of Reviews for Each Category (Sorted)')

# Add exact sum on top of each bar
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height, f'{height:.0f}', ha='center', va='bottom', rotation=90)

plt.tight_layout()
plt.show()
```

```python
mean_rating_by_category = df.groupby('Category')['Rating'].mean()
mean_rating_by_category
```

```python
# Set up the figure
plt.figure(figsize=(12, 6))
ax = plt.gca()

# Bar plot for mean rating
sns.barplot(x=mean_rating_by_category.index, y=mean_rating_by_category.values, ax=ax, color='blue', alpha=0.7)
ax.set_ylabel('Mean Rating')

# Scatter plot for sum of reviews
ax2 = ax.twinx()
ax2.scatter(x=sum_reviews_by_category.index, y=sum_reviews_by_category.values, color='red', s=100, label='Sum of Reviews')
ax2.set_ylabel('Sum of Reviews')
ax2.grid(False)

# Set the x-axis labels and rotate them for better visibility
ax.set_xticklabels(mean_rating_by_category.index, rotation=90)

# Show legend
ax2.legend(loc='upper right')

# Set title and display the plot
plt.title('Mean Rating and Sum of Reviews by Category')
plt.tight_layout()
plt.show()
```

```python
df['Rating Bin'] = pd.cut(df['Rating'], bins=10, precision=1)
df['Rating Bin']
```

```python
install_mean_by_rating = df.groupby('Rating Bin')['Installs (+)'].mean().reset_index()

# Bar Plot (Ratings vs. Mean Installs)
plt.figure(figsize=(10, 6))
sns.barplot(x='Rating Bin', y='Installs (+)', data=install_mean_by_rating, palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel('Rating')
plt.ylabel('Mean Installs (Log Scale)')
plt.yscale('log')  # Set y-axis to logarithmic scale
plt.title('Ratings vs. Mean Installs')
plt.show()
```
